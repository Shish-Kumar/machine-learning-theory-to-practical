# ***One-Hot Encoding***
- One-hot encoding is a technique used to convert categorical (text-based) data into numbers, so that machine learning models can understand it — because most ML algorithms only work with numbers, not text.
- In this technique, a new binary column (0 or 1) is created for every unique category. If a row belongs to that category, it gets a 1 in that column; all other category columns get 0.

### ***Which type of column is it applied on?***

> One-hot encoding is applied on **Nominal Categorical Columns** — columns where categories have **no inherent order/ranking**.

### From Example:

| Column | Type | Should you one-hot encode it? |
|--------|------|-------------------------------|
| **City** | Nominal | ✅ Yes — Hyderabad, Delhi, Mumbai have no ranking |
| **Department** | Nominal | ✅ Yes — IT, HR, Sales, Marketing have no ranking |
| **Employment_Type** | Nominal | ✅ Yes — Full-Time, Part-Time, Contract have no ranking |
| **Education_Level** | Ordinal | ❌ No — has natural order (High School → Diploma → Bachelor → Master → PhD). Use **Label/Ordinal Encoding** instead, so the model understands PhD > Master > Bachelor etc. |
| **Performance** | Target (Ordinal-like) | ❌ No — Average, Good, Excellent has order. Depending on the task, use **Ordinal Encoding** (for regression-style target) or **Label Encoding** (for classification target). |
| **Age**, **Monthly_Income** | Numerical | ❌ Not applicable — already numeric, no encoding needed |

### ***Why not one-hot encode Education_Level?***

> If you one-hot encode it, you'd get:

| Edu_HighSchool | Edu_Diploma | Edu_Bachelor | Edu_Master | Edu_PhD |
|----------------|-------------|--------------|------------|---------|
| 1 | 0 | 0 | 0 | 0 |

> This **destroys the order information** — the model won't know that PhD > Master > Bachelor. Instead, Ordinal Encoding should be used:

| Education_Level | Encoded Value |
|------------------|---------------|
| High School | 0 |
| Diploma | 1 |
| Bachelor | 2 |
| Master | 3 |
| PhD | 4 |

---

### Quick Rule of Thumb (applied to your dataset)

| Column Type | Your Columns | Encoding Technique |
|-------------|---------------|---------------------|
| Nominal (no order) | City, Department, Employment_Type | One-Hot Encoding |
| Ordinal (has order) | Education_Level, Performance | Label/Ordinal Encoding |
| Numerical | Age, Monthly_Income | No encoding needed |

> The **City** column is Nominal, so it's a perfect candidate for one-hot encoding:

| City |

# ***What is K-1 Encoding?***

- Suppose a categorical feature has **K unique categories**.
* **One-Hot Encoding** creates **K binary columns**.
* **K-1 Encoding** creates only **K-1 binary columns** by dropping one category.
- The dropped category becomes the **reference (baseline) category**.

#### ***Example***
> Original data:

| City      |
| --------- |
| Hyderabad |
| Delhi     |
| Mumbai    |

> Here, **K = 3**.

#### ***One-Hot Encoding (K Columns)***

| Hyderabad | Delhi | Mumbai |
| --------: | ----: | -----: |
|         1 |     0 |      0 |
|         0 |     1 |      0 |
|         0 |     0 |      1 |

#### ***K-1 Encoding***

> Suppose we drop **Mumbai**.

| Hyderabad | Delhi |
| --------: | ----: |
|         1 |     0 |
|         0 |     1 |
|         0 |     0 |

> When both columns are **0**, it automatically means the city is **Mumbai**.

---

#### ***Why Do We Use K-1 Encoding?***

> With K columns, one column can always be determined from the others.

-  For example:

```text
Mumbai = 1 − (Hyderabad + Delhi)
```

> This creates **redundant information**, causing **multicollinearity** in some machine learning models, especially **Linear Regression** and **Logistic Regression**.
- To avoid this, we remove one category.
