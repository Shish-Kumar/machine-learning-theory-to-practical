# ***Manual Lable Encoding***

```python
# Create a Mapping Dictionary
# Assign a unique label to each category.

performance_mapping = {
    "Average": 0,
    "Good": 1,
    "Excellent": 2
}
```

```python
# Apply the Mapping
df["Performance"] = df["Performance"].map(performance_mapping)
```

```python
df['Performance'].unique()
```

# ***Lable Encoding using LableEncoder  Class***
- always apply lable encoder after train test split to prevent data Leakage, here we are just learnign the concept.

#### ***Reload the dataset first because we have already encoded the Performance column manually.***

```python
df  = pd.read_excel('ML_Preprocessing_Practice_Dataset.excel')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["Performance"] = le.fit_transform(df['Performance'])
```

```python
df['Performance']
```

```python
df.head()
```

```python
le.classes_
```

```python
# This confirms that LabelEncoder assigns labels in alphabetical order:
Average   → 0
Excellent → 1
Good      → 2
```

### ***You can also convert the encoded values back to their original categories:***

```python
le.inverse_transform([0, 1, 2])
```
