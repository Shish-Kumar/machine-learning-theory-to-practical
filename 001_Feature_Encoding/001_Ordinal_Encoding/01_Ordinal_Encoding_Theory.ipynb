# ***Ordinal Encoding***
> **Ordinal Encoding** is a feature encoding technique used to convert **ordinal categorical data** into numerical values while preserving the natural order or ranking of the categories.

- **Short Interview Definition**
> **Ordinal Encoding is the process of converting ordinal categorical values into numerical values while maintaining their natural order.**

---

###  ***Why Do We Need Ordinal Encoding?***

- Machine Learning algorithms work with **numerical data**, not text.
- If a feature contains text values such as:

| Education_Level |
| --------------- |
| High School     |
| Diploma         |
| Bachelor        |
| Master          |
| PhD             |

- the model cannot understand these text values directly.

- Therefore, we convert them into numbers.

- Example:

| Education_Level | Encoded Value |
| --------------- | ------------: |
| High School     |             0 |
| Diploma         |             1 |
| Bachelor        |             2 |
| Master          |             3 |
| PhD             |             4 |

- Notice that the **order is preserved**.

---

###   ***When Should We Use Ordinal Encoding?***

> Ordinal Encoding should be used when the categories have a **natural order or ranking**.

- Examples:

| Feature                                                | Order Exists? | Use Ordinal Encoding? |
| ------------------------------------------------------ | ------------- | --------------------- |
| Education Level                                        |  Yes         |  Yes                 |
| T-Shirt Size (S, M, L, XL)                             |  Yes         |  Yes                 |
| Customer Satisfaction (Poor, Average, Good, Excellent) |  Yes         |  Yes                 |
| Movie Rating (1–5 Stars)                               |  Yes         |  Yes                 |

---

###  ***When Should We NOT Use Ordinal Encoding?***

> If the categories **do not have any natural order**, do not use Ordinal Encoding.

Example:

| City      |
| --------- |
| Delhi     |
| Mumbai    |
| Hyderabad |
| Chennai   |

> There is no ranking among cities.

- Similarly,

| Color |
| ----- |
| Red   |
| Blue  |
| Green |

> There is no natural order.
> For such features, we use **One-Hot Encoding**.

---

# ***Types of Encoding***
> There are several encoding techniques used in Machine Learning.

| Encoding Technique     | Used For                                                     |
| ---------------------- | ------------------------------------------------------------ |
| **Ordinal Encoding**   | Ordinal categorical features (ordered categories)            |
| **One-Hot Encoding**   | Nominal categorical features (unordered categories)          |
| **Label Encoding**     | Mainly for the target (dependent) variable in classification |
| **Target Encoding**    | High-cardinality categorical features                        |
| **Binary Encoding**    | High-cardinality categorical features                        |
| **Frequency Encoding** | Categories replaced with their frequency/count               |

> In this course, we will first learn the three most commonly used techniques:
>
> * Ordinal Encoding
> * One-Hot Encoding
> * Label Encoding

# ***How Does Ordinal Encoding Work?***
> It assigns a unique integer to each category according to its order.

- Example

| Education_Level | Encoded Value |
| --------------- | ------------: |
| High School     |             0 |
| Diploma         |             1 |
| Bachelor        |             2 |
| Master          |             3 |
| PhD             |             4 |

> The numerical values represent the **rank**, not the magnitude.

> For example, assigning `Master = 3` and `Bachelor = 2` indicates that **Master is ranked higher than Bachelor**. It does **not** mean that a Master's degree is "1 unit greater" in a mathematical sense.

---


 This is one of the most common misconceptions.
###  ***`OrdinalEncoder` does NOT automatically detect the correct natural order.***

> It only knows the order that **you provide**, or if you don't provide one, it uses its own default ordering (typically lexicographical/alphabetical), which may not match the real-world ranking.

### Example

- Suppose your data is:

| Education_Level |
| --------------- |
| Bachelor        |
| PhD             |
| Master          |
| Diploma         |
| High School     |

> If you simply write:

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

df["Education_Level"] = encoder.fit_transform(df[["Education_Level"]])
```

> The encoder may assign values based on alphabetical order:

| Category    | Encoding |
| ----------- | -------: |
| Bachelor    |        0 |
| Diploma     |        1 |
| High School |        2 |
| Master      |        3 |
| PhD         |        4 |

> This is **alphabetical order**, not necessarily the intended educational hierarchy.

---

### Correct Way

> You should explicitly specify the order:

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(categories=[[
    "High School",
    "Diploma",
    "Bachelor",
    "Master",
    "PhD"
]])

df["Education_Level"] = encoder.fit_transform(df[["Education_Level"]])
```

> Now the encoding becomes:

| Category    | Encoding |
| ----------- | -------: |
| High School |        0 |
| Diploma     |        1 |
| Bachelor    |        2 |
| Master      |        3 |
| PhD         |        4 |

---

### ***Interview Answer***

- **Q: Does `OrdinalEncoder` automatically detect the correct order?**

- **Answer:**

> **No. `OrdinalEncoder` does not understand the real-world meaning of categories. It uses the order provided by the user. If no order is specified, it uses a default ordering (typically alphabetical), which may not represent the true ordinal relationship.**
