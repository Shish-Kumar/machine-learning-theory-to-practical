# Pipeline in Scikit-learn

## Definition

A **Pipeline** is a Scikit-learn object that **connects multiple machine learning steps into a single workflow**, where the output of one step becomes the input of the next step.

It allows preprocessing, feature engineering, and model training to be executed in the correct order automatically.

**Simple Definition (Interview):**

> **A Pipeline is a sequence of data processing and machine learning steps that are executed automatically in the correct order as one single object.**

---

# Why was Pipeline introduced?

In a machine learning project, we rarely train a model directly on raw data.

Usually, the workflow looks like this:

```
Raw Data
     ↓
Missing Value Handling
     ↓
Encoding
     ↓
Feature Scaling
     ↓
Feature Selection (Optional)
     ↓
Model Training
     ↓
Prediction
```

If every step is written separately, the code becomes:

* lengthy
* difficult to maintain
* easy to make mistakes
* difficult to deploy

Pipeline solves these problems by combining all these steps into one object.

---

# Real-Life Example

Imagine making coffee.

```
Coffee Beans
      ↓
Grind
      ↓
Boil Water
      ↓
Mix Coffee
      ↓
Add Milk
      ↓
Serve
```

Each step depends on the previous one.

You cannot add milk before grinding the coffee.

Similarly, in Machine Learning:

```
Dataset
     ↓
Clean Data
     ↓
Scale Features
     ↓
Train Model
     ↓
Predict
```

Each step must happen in the correct sequence.

A Pipeline ensures that sequence is always followed.

---

# What happens inside a Pipeline?

Suppose we create:

```python
Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])
```

During training (`fit`):

```
X_train
    ↓
StandardScaler.fit()
    ↓
StandardScaler.transform()
    ↓
LogisticRegression.fit()
```

During prediction (`predict`):

```
X_test
    ↓
StandardScaler.transform()
    ↓
LogisticRegression.predict()
```

Notice something important:

During prediction,

```
Scaler DOES NOT fit again.
```

It only transforms the new data using the statistics learned from the training data.

This prevents **data leakage**.

---

# Components of a Pipeline

A Pipeline contains two types of steps.

## 1. Transformers

A transformer changes the data.

Examples:

* StandardScaler
* MinMaxScaler
* SimpleImputer
* OneHotEncoder
* PCA

They usually have:

```
fit()
transform()
fit_transform()
```

Example:

```
Age

20
40
60

↓

StandardScaler

↓

-1.2
0
1.2
```

The data changes.

---

## 2. Estimator

The last step is usually an estimator.

It trains the model.

Examples:

* LinearRegression
* LogisticRegression
* RandomForestClassifier
* DecisionTreeClassifier
* SVC

Estimators usually have:

```
fit()

predict()
```

---

# Pipeline Structure

```
Raw Data
      ↓
Transformer
      ↓
Transformer
      ↓
Transformer
      ↓
Estimator
```

Only the **last step** is generally the machine learning model.

All previous steps are transformers.

---

# Example Pipeline

```
Dataset
      ↓
SimpleImputer
      ↓
StandardScaler
      ↓
PCA
      ↓
LogisticRegression
```

Step by step:

```
Missing values filled

↓

Features scaled

↓

Dimensions reduced

↓

Model trained
```

# Code Example

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])

pipe.fit(X_train, y_train)

predictions = pipe.predict(X_test)
```

The pipeline automatically performs:

```
Scale data

↓

Train model

↓

Predict
```

You do not need to manually scale the test data.

---

# Advantages of Pipeline

### 1. Cleaner Code

Without Pipeline

```
Scale

↓

Train

↓

Predict
```

Many separate lines.

With Pipeline

```
One object

↓

One fit()

↓

One predict()
```

---

### 2. Prevents Data Leakage

The scaler learns only from the training data.

During prediction, it only transforms the new data.

This makes evaluation more reliable.

---

### 3. Easy Deployment

Instead of saving multiple objects:

```
Scaler.pkl

Model.pkl
```

You can save one object:

```
pipeline.pkl
```

When loaded later, it automatically performs preprocessing and prediction.

---

### 4. Works Well with GridSearchCV and Cross-Validation

The entire workflow is validated together.

You don't have to preprocess separately before every fold.

---

# Key Rules

* Steps are executed **from top to bottom**.
* Every step except the last should be a **transformer**.
* The last step is usually an **estimator (model)**.
* During training, transformers use `fit()` and `transform()`, then the model uses `fit()`.
* During prediction, transformers use only `transform()`, then the model uses `predict()`.

---

# Interview Definition

> **A Pipeline in Scikit-learn is a sequence of preprocessing and machine learning steps combined into a single object. It ensures that each step is executed automatically in the correct order, making the workflow cleaner, reusable, and less prone to errors such as data leakage.**
