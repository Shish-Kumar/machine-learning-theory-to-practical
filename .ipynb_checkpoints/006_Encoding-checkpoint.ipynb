# ***1. LableEncoder***


---

### ***Label Encoding – Important Points***

1. **Definition**

   * Categorical (string/text) data ko numeric (integer) values me convert karne ki technique.
   * Har unique category → ek unique integer.

2. **Working**

   * Example:

     ```
     Color: [Red, Blue, Green, Blue]  
     Encoded: [2, 0, 1, 0]
     ```
   * Mapping usually **alphabetical order** ya library ke internal order par hota hai.

3. **Where to Use**

   * Jab categories **nominal** (no order) ho aur aap **tree-based models** use kar rahe ho (Decision Tree, Random Forest, XGBoost).
   * Jab categories ka number bahut zyada na ho.

4. **Limitations**

   * **Artificial order create hota hai** (0 < 1 < 2), jo galat interpretation de sakta hai.
   * Linear models (Regression, Logistic Regression, KNN, SVM) ke liye confusion create kar sakta hai.
   * Agar data ordinal ho (Small < Medium < Large), to **Ordinal Encoding** better hai.

5. **Advantages**

   * Simple aur fast.
   * Memory efficient (sirf ek column).
   * Tree-based models ke liye useful.

6. **Alternatives**

   * **One-Hot Encoding** → jab categories ka order na ho aur linear models use kar rahe ho.
   * **Ordinal Encoding** → jab categories me natural order ho.

---

👉 Ek line me:
**Label Encoding** = *Category ko ek number ID dena — fast aur simple, par caution with linear models.*

---

````python
from sklearn.preprocessing import LabelEncoder
visa_data  = pd.read_csv('Visadataset.csv')
le = LabelEncoder()

visa_data['case_status']= le.fit_transform(visa_data['case_status'])
visa_data.head(3)
````

# ***2. Ordinal Encoding***



### ***Ordinal Encoding – Important Points***

1. **Definition**  
   - Categorical variables ko numbers me convert karta hai, **lekin natural order ko preserve karta hai**.  
   - Har category ko manually ya predefined order ke according integer diya jata hai.  

2. **Working**  
   - Example:  
     ```
     Size: [Small, Medium, Large]  
     Ordinal Encoding: Small=0, Medium=1, Large=2
     ```

3. **Where to Use**  
   - Jab categorical data **ordinal type** ho (jisme rank/order ho).  
   - Useful for **linear models** (Regression, Logistic Regression, SVM) jo order ka sense samajh pate hain.  

4. **Limitations**  
   - **Nominal data ke liye use nahi karna chahiye** (jaise colors, cities), warna model galat order assume karega.  
   - Agar categories ka order sahi define na kiya jaye to model mislead ho sakta hai.  

5. **Advantages**  
   - Order information ko preserve karta hai.  
   - Simple aur memory-efficient (sirf ek column banta hai, one-hot ki tarah dimension nahi badhata).  

6. **Alternatives**  
   - **Label Encoding** → jab nominal data ho aur tree-based models use kar rahe ho.  
   - **One-Hot Encoding** → jab order nahi ho aur linear models use karna ho.  

---

👉 Ek line me:  
**Ordinal Encoding** = *Ordered categories ko number me convert karna — best for ordinal features with meaningful ranking.*  

---


````python
visa_data  = pd.read_csv('Visadataset.csv')
cat = visa_data.select_dtypes(include='object').columns
num = visa_data.select_dtypes(exclude='object').columns
````

````python
visa_data.head()
````

````python
import numpy as np
from sklearn.preprocessing import OrdinalEncoder

# Sample data
data = np.array([
    ["Small", "High School"],
    ["Medium", "Bachelor"],
    ["Large", "Master"],
    ["Medium", "PhD"],
    ["Small", "Bachelor"]
])

# Define orders for each column
size_order = ["Small", "Medium", "Large"]
edu_order = ["High School", "Bachelor", "Master", "PhD"]

encoder = OrdinalEncoder(categories=[size_order, edu_order])

encoded = encoder.fit_transform(data)

data,encoded
````

---

## 🔹 OrdinalEncoder ke do mode hote hain:

### 1. ***Without passing order (default mode)***

```python
Oe = OrdinalEncoder()
visa_data['education_encoded'] = Oe.fit_transform(visa_data[['education_of_employee']])
```

👉 Is case me sklearn **khud order decide karta hai**, aur wo alphabetically sort kar deta hai.

For example:

```
["Bachelor's", 'Doctorate', 'High School', "Master's"]
```

Ye sequence banega, aur encoding hogi:

```
Bachelor's  → 0  
Doctorate   → 1  
High School → 2  
Master's    → 3  
```

⚠️ Problem: Education me natural hierarchy hoti hai, par alphabetical order usko follow nahi karta.

---

### 2. ***With manual order (custom mode)***

```python
education_order = [['High School', "Bachelor's", "Master's", "Doctorate"]]
Oe = OrdinalEncoder(categories=education_order)
visa_data['education_encoded'] = Oe.fit_transform(visa_data[['education_of_employee']])
```

👉 Ab encoding hogi exactly waise jaise aapne order diya hai:

```
High School → 0  
Bachelor's  → 1  
Master's    → 2  
Doctorate   → 3  
```

✅ Ye sahi hai, kyunki education ek **ordinal variable** hai jisme hierarchy aap define karte ho, alphabetically nahi.

---

## 🔑 Final Baat

* Agar aap **nominal categorical data** encode kar rahe ho → `OrdinalEncoder` ko order dene ki zarurat nahi, alphabetical chalega (waise is case me OneHot zyada use hota hai).
* Agar aap **ordinal categorical data** encode kar rahe ho → aapko **apna order pass karna hi padega**, warna sklearn alphabetic order le lega jo galat hoga.

---

👉 Matlab: `OrdinalEncoder` automatic to hai, par wo **domain knowledge** nahi samajhta. Isliye natural hierarchy hamesha aapko batani padti hai.

# ***2. OneHotEncoder***
-  the method name is : pd.getdummies


---

###  ***One-Hot Encoding – Important Points***

1. **Definition**

   * One-Hot Encoding ek technique hai jo **categorical (nominal/ordinal) features** ko numeric form me convert karti hai.
   * Har category ke liye ek **new binary column (0/1)** banaya jaata hai.

---

2. **Working**

   * Suppose feature `"Color"` = {Red, Blue, Green}.
   * One-Hot Encoding ke baad:

     ```
     Red   → [1,0,0]
     Blue  → [0,1,0]
     Green → [0,0,1]
     ```

---

3. **Key Properties**

   * Order maintain **nahi** karta (unlike label/ordinal encoding).
   * Suitable for **nominal data** (jisme koi hierarchy/order nahi hota).
   * Linear models aur distance-based models (Logistic Regression, SVM, KNN) ke liye best, kyunki ye dummy variables create karta hai.

---

4. **Advantages**
   ✅ Koi **false order** introduce nahi hota.
   ✅ Models ko categories clearly distinguish karne me help karta hai.
   ✅ Simple aur widely supported.

---

5. **Disadvantages**
   ❌ Agar categories bahut zyada ho (high cardinality), to feature space **dimensionality explosion** ho jaata hai.
   ❌ Memory usage badh jaata hai (sparse matrix).
   ❌ Tree-based models me (Decision Tree, Random Forest) kabhi-kabhi unnecessary hota hai, kyunki wo label encoding se bhi sahi kaam kar lete hain.

---

6. **When to Use**

   * Nominal categorical features (Gender, Color, City, Country, etc.).
   * Linear/Distance-based ML models.
   * Jab categories **limited** aur **manageable** ho.

---

7. **When NOT to Use**

   * High-cardinality features (e.g., Zipcode, UserID, ProductID).
   * Tab better hai **Target Encoding / Frequency Encoding** use karna.

---

📌 **Shortcut yaad rakhne ka:**

* **Nominal = One-Hot** (no order, separate columns).
* **Ordinal = Ordinal Encoding** (has order, numeric rank).

---


````python
import pandas as pd

data = {
    'Employee': ['A', 'B', 'C', 'D'],
    'Department': ['HR', 'Finance', 'IT', 'HR']
}

df = pd.DataFrame(data)
df
````

### ***One-Hot Encoding with pandas***

````python
df_encoded = pd.get_dummies(df, columns=['Department'])
df_encoded
````

- Default: sparse_output=True → sparse matrix मिलता है।
- अगर तुम्हें normal array चाहिए → sparse_output=False कर दो।

````python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Encoder initialize karo - sirf sparse_output use karo
ohe = OneHotEncoder(sparse_output=False)  # Dense array ke liye

# Fit aur transform karo
encoded = ohe.fit_transform(df[['Department']])

# DataFrame me convert karo
df_ohe = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(['Department']))
df_ohe
````

## ***LableEncoder vs OneHotEncoder***
- generally we dont want a reletion b\w input columns
- we neeed both input columns are not correlated each other
- we need both input columns are `independent (X,Y)` each other
- we need both input columns are `perpendicular` each other
- we need both input columns are `orthogonal` each other
- we need both input columns are `90 degree phase` shift each other
- OneHotEncoder will achive avobe

    - columns are increases means dimenssions are increase

चलो `sklearn.preprocessing.OneHotEncoder` के arguments को आसान हिंदी में समझते हैं, और हर argument के साथ एक छोटा example भी दूँगा।

---

## 🔹 `categories`

* **क्या करता है:** तय करता है कि encoder किस set of categories (unique values) को use करेगा।
* Default: `'auto'` → मतलब ये खुद dataset से unique values ले लेगा।
* आप manually भी categories define कर सकते हो।

**Example:**

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

X = np.array([['Red'], ['Green'], ['Blue']])

# Auto (default) → categories खुद ले लेगा
enc = OneHotEncoder(categories='auto')
print(enc.fit(X).categories_)  
# [['Blue', 'Green', 'Red']]

# Manual
enc = OneHotEncoder(categories=[['Red','Green','Blue']])
print(enc.fit(X).categories_)  
# [['Red', 'Green', 'Blue']]
```

---

## 🔹 `drop`

* **क्या करता है:** Dummy variable trap से बचने के लिए किसी एक category को drop कर सकता है।
* Options:

  * `None` → कुछ भी drop नहीं होगा (default)
  * `'first'` → हर column से पहली category drop होगी।
  * `'if_binary'` → अगर column binary है, तो एक category drop कर देगा।
  * `['category_name']` → manually भी specify कर सकते हैं।

**Example:**

```python
X = np.array([['Male'], ['Female'], ['Male']])

enc = OneHotEncoder(drop='if_binary')
print(enc.fit_transform(X).toarray())
# केवल 1 कॉलम मिलेगा (Male या Female में से एक drop हो जाएगा)
```

---

## 🔹 `sparse_output`

* **क्या करता है:** Output matrix को sparse (compressed) या dense (normal numpy array) बनाएगा।
* Default: `True` → sparse matrix return करता है।
* अगर आप dense चाहते हो तो `sparse_output=False` कर दो।

**Example:**

```python
X = np.array([['A'], ['B'], ['C']])

enc = OneHotEncoder(sparse_output=False)
print(enc.fit_transform(X))
# Dense numpy array output मिलेगा
```

---

## 🔹 `dtype`

* **क्या करता है:** Output matrix का data type set करता है।
* Default: `np.float64`
* आप चाहो तो `int` भी कर सकते हो।

**Example:**

```python
X = np.array([['Yes'], ['No']])

enc = OneHotEncoder(dtype=int, sparse_output=False)
print(enc.fit_transform(X))
# [[0 1]
#  [1 0]]
```

---

## 🔹 `handle_unknown`

* **क्या करता है:** Unknown category (जो training में नहीं थी) आने पर क्या करना है।
* Options:

  * `'error'` (default) → error देगा
  * `'ignore'` → उस category को ignore करेगा और पूरा row zeros से भर देगा।
  * `'infrequent_if_exist'` → rare categories को "infrequent" bucket में डाल देगा।

**Example:**

```python
X_train = np.array([['Cat'], ['Dog']])
X_test  = np.array([['Horse']])  # Unknown category

enc = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
enc.fit(X_train)
print(enc.transform(X_test))
# [[0. 0.]]  # unknown category → सब zero
```

---

## 🔹 `min_frequency`

* **क्या करता है:** Rare categories (बहुत कम बार आने वाली) को combine कर देगा।
* Value: float (fraction) या int (absolute count)।

**Example:**

```python
X = np.array([['A'], ['A'], ['B'], ['C'], ['D']])

enc = OneHotEncoder(min_frequency=2, sparse_output=False)
print(enc.fit_transform(X))
# Rare categories को एक 'infrequent' category में merge कर देगा।
```

---

## 🔹 `max_categories`

* **क्या करता है:** किसी feature के लिए maximum categories limit कर देता है।
* सबसे frequent categories ही रखेगा, बाकी को "infrequent" bucket में डाल देगा।

**Example:**

```python
X = np.array([['A'], ['B'], ['C'], ['D'], ['E'], ['F'], ['A']])

enc = OneHotEncoder(max_categories=3, sparse_output=False)
print(enc.fit_transform(X))
# केवल top 3 categories one-hot होंगी, बाकी "infrequent" bucket में
```

---

## 🔹 `feature_name_combiner`

* **क्या करता है:** Output feature names कैसे दिखेंगे, ये decide करता है।
* Default: `'concat'` → feature name और category को जोड़ देगा।
* Option: `'indicator'` → केवल category का नाम दिखेगा।

**Example:**

```python
X = np.array([['Red'], ['Green']])

enc = OneHotEncoder(sparse_output=False, feature_name_combiner='concat')
enc.fit(X)
print(enc.get_feature_names_out())
# ['x0_Blue' 'x0_Green' 'x0_Red']
```
