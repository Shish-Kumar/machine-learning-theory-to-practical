# ***ColumnTransformer***
* `Definition` : ColumnTransformer is a Scikit-learn class used to apply different preprocessing transformations to different columns of a dataset and combine the transformed results into a single feature matrix.

* `Simple Definition` : ColumnTransformer allows us to preprocess numerical and categorical columns differently in one step before training a machine learning model.

#### ***Why do we need ColumnTransformer?***
- In a real dataset, different columns require different preprocessing.

| Age | Salary | Gender | City   |
| --- | ------ | ------ | ------ |
| 22  | 30000  | Male   | Delhi  |
| 35  | 60000  | Female | Mumbai |


>  Age and Salary are numerical → Scale them.

> Gender and City are categorical → Encode them.

- Instead of preprocessing each column separately, ColumnTransformer does everything together.

### ***Syntax***

````python
from sklearn.compose import ColumnTransformer

ColumnTransformer(
    transformers=[
        ('name', transformer, columns)
    ],
    remainder='drop',
    sparse_threshold=0.3,
    n_jobs=None,
    verbose_feature_names_out=True
)
````

### ***Parameters***

#### 1. transformers (Required)
> A list of tuples that defines which transformation to apply.

#### Syntax

```python
transformers=[
    ('name', transformer, columns)
]
```

> Each tuple contains:

* Name
* Transformer
* Columns

> Example

```python
transformers=[
    ('num', StandardScaler(), ['Age', 'Salary']),
    ('cat', OneHotEncoder(), ['Gender', 'City'])
]
```

---

#### 2. Name

> A user-defined name for the transformer.

```python
('num', ...)
```

- or

```python
('cat', ...)
```

> It is only an identifier.

---

#### 3. Transformer

> The preprocessing object to apply.

- Examples

```python
StandardScaler()

MinMaxScaler()

OneHotEncoder()

OrdinalEncoder()

SimpleImputer()
```

- Example

```python
('num', StandardScaler(), ['Age'])
```

---

#### 4. Columns

> Specify which columns should receive the transformation.

- Using column names

```python
['Age', 'Salary']
```

- Using column index

```python
[0, 1]
```

---

#### 5. remainder

> Controls what happens to columns not listed in `transformers`.

### Default

```python
remainder='drop'
```

- Unspecified columns are removed.

### Keep remaining columns

```python
remainder='passthrough'
```

- Unspecified columns remain unchanged.

---

#### 6. sparse_threshold

> Determines whether the output should be sparse or dense.

- Default

```python
0.3
```

- Usually you don't change it.

---

#### 7. n_jobs

> Runs transformations in parallel.

```python
n_jobs=-1
```

- Uses all CPU cores.

---

#### 8. verbose_feature_names_out

> Controls output feature names.

- Default

```python
True
```

- Output

```
num__Age
cat__Gender_Male
```

- False

```python
False
```

- Output

```
Age
Gender_Male
```

---

#### Working Flow

```
Dataset
    │
    ▼
ColumnTransformer
    │
    ├──────── Numerical Columns
    │              │
    │              ▼
    │        StandardScaler
    │
    ├──────── Categorical Columns
    │              │
    │              ▼
    │        OneHotEncoder
    │
    ▼
Combined Features
    │
    ▼
Machine Learning Model
```

#### Example

> Suppose

```text
Age  Salary  Gender  City

22   30000   Male    Delhi
35   60000   Female  Mumbai
28   45000   Male    Hyderabad
```

> We want

* Scale Age and Salary
* Encode Gender and City

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Age', 'Salary']),
        ('cat', OneHotEncoder(), ['Gender', 'City'])
    ]
)

X_transformed = ct.fit_transform(X)
```

### ***What happens internally?***

#### Step 1

- Numerical columns

```
Age
Salary
```

- ↓

```
StandardScaler
```

- ↓

```
Scaled Age
Scaled Salary
```

---

#### Step 2

> Categorical columns

```
Gender
City
```

- ↓

```
OneHotEncoder
```

- ↓

```
Gender_Male
Gender_Female
City_Delhi
City_Mumbai
City_Hyderabad
```

---

#### Step 3

> Both outputs are combined into one matrix.

```
Scaled Age
Scaled Salary
Gender_Male
Gender_Female
City_Delhi
City_Mumbai
City_Hyderabad
```
