# ***Label Encoding***

- **Label Encoding** is a feature encoding technique that converts categorical values into numerical labels by assigning a unique integer to each category.

### ***Short Interview Definition***

> **Label Encoding converts categorical values into unique numerical labels.**

---

# Why Do We Need Label Encoding?

Machine Learning algorithms cannot work directly with text (categorical) values.

For example,

| Performance |
| ----------- |
| Good        |
| Excellent   |
| Average     |

> Since these are text values, they must be converted into numbers before training a machine learning model.

> After Label Encoding,

| Performance | Label |
| ----------- | ----: |
| Average     |     0 |
| Excellent   |     1 |
| Good        |     2 |

> Each category receives a unique integer label.

---

# When Should We Use Label Encoding?

### ***Label Encoding is **primarily used for the target (dependent) variable** in classification problems.***

- Examples:

> * Spam / Not Spam
> * Positive / Negative
> * Pass / Fail
> * Yes / No
> * Cat / Dog / Horse
> * Employee Performance (Average, Good, Excellent)

---

###  ***Should We Use Label Encoding for Input Features?***

> Generally, **No**.
> If an input feature is **nominal** (no natural order), Label Encoding should not be used because the model may incorrectly assume an order between the categories.


> Example:

| City      |
| --------- |
| Delhi     |
| Mumbai    |
| Hyderabad |

> After Label Encoding:

| City      | Label |
| --------- | ----: |
| Delhi     |     0 |
| Mumbai    |     1 |
| Hyderabad |     2 |

> This may incorrectly suggest:

```text
Hyderabad > Mumbai > Delhi
```

> However, no such ranking exists.
> For nominal input features, **One-Hot Encoding** is usually the better choice.

---

### ***Can Label Encoding Be Used for Ordinal Features?****

>  Although it is technically possible, **Ordinal Encoding** is the preferred technique because it allows you to explicitly define the correct order of the categories.

---

### ***How Does Label Encoding Work?***

> It assigns a unique integer to every unique category.

> Example:

| Performance | Label |
| ----------- | ----: |
| Average     |     0 |
| Good        |     1 |
| Excellent   |     2 |

> The numbers simply act as labels. They are identifiers and should not be interpreted as mathematical values.

---

# ***Label Encoding vs Ordinal Encoding***

| Label Encoding                      | Ordinal Encoding                        |
| ----------------------------------- | --------------------------------------- |
| Assigns unique labels to categories | Preserves a predefined order            |
| Mostly used for the target variable | Used for ordinal input features         |
| Order is not specified by the user  | User defines the correct category order |
| Common in classification problems   | Common for ordered categorical features |

---

# Interview Questions

### What is Label Encoding?

> Label Encoding converts categorical values into unique numerical labels.

### When should Label Encoding be used?

> It is primarily used for the target (dependent) variable in classification problems.

### Why is Label Encoding generally not recommended for nominal input features?

> Because it may introduce an artificial order between categories, which can mislead certain machine learning algorithms.

### ***How Does Label Encoding Assign Numbers?***

>Unlike **Ordinal Encoding**, **Label Encoding does not allow us to define the order of the categories**.

>Instead, it **automatically assigns numerical labels based on the alphabetical (lexicographical) order** of the categories.

- For example,

| Performance |
| ----------- |
| Good        |
| Excellent   |
| Average     |

> After applying `LabelEncoder`, the output becomes:

| Category  | Label |
| --------- | ----: |
| Average   |     0 |
| Excellent |     1 |
| Good      |     2 |

> Notice that the labels are assigned in **alphabetical order**, not based on the meaning or importance of the categories.

---

### ***Important Note***

>Unlike **OrdinalEncoder**, **LabelEncoder does not provide an option to manually specify the category order**.

>The encoder always assigns labels according to the alphabetical order of the unique values.

---

### ***Interview Question***

> **Q. On what basis does `LabelEncoder` assign numerical labels?**

- **Answer:**

> **`LabelEncoder` assigns labels based on the alphabetical (lexicographical) order of the unique categories. The user cannot manually specify the order.**



## ***Q. Does the order assigned by LabelEncoder affect the target variable?*** 🟢

- Answer:

>Generally, no. In classification problems, the target values are treated as class labels, not as ordered numerical values. Therefore, the alphabetical order assigned by LabelEncoder usually does not affect model training. However, for input features, using LabelEncoder on nominal data can introduce a false order and should be avoided.

# Methods

## 1. `fit()`

Learns all the unique categories present in the data.

### Syntax

```python
encoder.fit(data)
```

### Example

```python
encoder.fit(df["Performance"])
```

After calling `fit()`, the encoder stores all unique categories internally.

---

## 2. `transform()`

Converts the learned categories into numerical labels.

### Syntax

```python
encoder.transform(data)
```

### Example

```python
encoder.transform(df["Performance"])
```

Example Output

```text
[1 2 0 1 2]
```

---

## 3. `fit_transform()`

Performs both operations together.

* Learns the categories.
* Converts them into numerical labels.

### Syntax

```python
encoder.fit_transform(data)
```

### Example

```python
df["Performance"] = encoder.fit_transform(df["Performance"])
```

This is the most commonly used method.

---

## 4. `inverse_transform()`

Converts encoded labels back to their original categories.

### Syntax

```python
encoder.inverse_transform(encoded_data)
```

### Example

```python
encoder.inverse_transform([0, 1, 2])
```

Example Output

```text
['Average' 'Excellent' 'Good']
```

---

# Important Attributes

## `classes_`

Displays the categories learned by the encoder.

### Example

```python
encoder.classes_
```

Example Output

```text
array(['Average', 'Excellent', 'Good'], dtype=object)
```

Notice that the categories are arranged in **alphabetical order**.

# Does `LabelEncoder` Have Parameters?

Unlike `SimpleImputer` or `OneHotEncoder`, **`LabelEncoder` does not have commonly used initialization parameters**.

You simply create the object:

```python
encoder = LabelEncoder()
```

and then use its methods (`fit()`, `transform()`, `fit_transform()`, and `inverse_transform()`).
